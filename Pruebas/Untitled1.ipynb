```python
%matplotlib inline
import cv2
import matplotlib
from matplotlib import colors
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = cv2.imread('C:\\Users\\Cristhian\\TEETH-RECOGNITION-WITH-MACHINE-LEARNING\\DATASET - Recortado\\101_0011.JPG')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image = cv2.resize(image,None,fx=1/3,fy=1/3)
# Convert from RGB to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV_FULL)

images = []
for i in [0, 1, 2]:
    colour = hsv.copy()
    if i != 0: colour[:,:,0] = 0
    if i != 1: colour[:,:,1] = 255
    if i != 2: colour[:,:,2] = 255
    images.append(colour)

hsv_stack = np.vstack(images)
rgb_stack = cv2.cvtColor(hsv_stack, cv2.COLOR_HSV2RGB_FULL)
plt.imshow(rgb_stack)

#flags = [i for i in dir(cv2) if i.startswith('COLOR_')]
#print (flags)
```
